# Seaborn reference

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
```

```python
df = sns.load_dataset("penguins")
g = sns.scatterplot(df, x="bill_length_mm", y="flipper_length_mm", hue="species")
g.set_xlabel("Bill length (mm)")
g.set_ylabel("Flipper Length (mm)")
g.legend_.set_title("Penguin species")
```

```python
g = sns.pairplot(df, hue='species')
g.
```

```python
g = sns.countplot(df, x="species", hue="species")
```

```python
g = sns.barplot(df, x='sex', y='bill_length_mm', hue='species')
sns.move_legend(g, 'lower center', bbox_to_anchor=(0.5,1), ncol=3, frameon=False)
```

```python
g = sns.histplot(df, x='bill_length_mm', hue='species', multiple='stack')
```

```python
g = sns.kdeplot(df, x='bill_length_mm', hue='species', fill=True, alpha=0.6)
```

### Facet-grid density plot
`height` controls the height of a single facet, while `aspect` controls the width as a multiplier of height. 

```python
g = sns.FacetGrid(df, row='species', hue='species', margin_titles=True, aspect=4, height=2)
g.map_dataframe(sns.kdeplot, x="bill_length_mm", fill=True)
g.set_titles(row_template="{row_name}", position=[.9, .9])
g.set_axis_labels(y_var="Density", x_var="Bill length (mm)")

```

```python
g = sns.displot(df, row='species', x='bill_length_mm', hue='species', kind='kde', fill=True, aspect=4, height=2, legend=False)
g.set_titles(row_template="{row_name}", position=[0.95, 1])
g.set_axis_labels(y_var="Density", x_var="Bill length (mm)")
```

```python
g = sns.boxplot(df, x='sex', y='bill_length_mm', hue='species')
sns.move_legend(g, 'lower center', bbox_to_anchor=(0.5,1), ncol=3, frameon=False)
```
